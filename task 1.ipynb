```python
# Example: Download Apple (AAPL) stock data from Jan 2023 to Jan 2024
data = yf.download('AAPL', start='2023-01-01', end='2024-01-01')

print(data.head())  # Shows the first 5 rows
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.plot(data['Close'], label='Apple Stock Price', color='blue')
plt.title('Apple Stock Price (AAPL)')
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.legend()
plt.show()
```

```python
# Example: Apple (AAPL) stock for last 1 year
stock = yf.download('AAPL', start='2023-07-01', end='2024-07-01')

# Preview the data
print(stock.head())
```

```python
print(stock.columns.tolist())
```

```python
stock['Daily Return'] = stock['Close'].pct_change()
print(stock[['Close', 'Daily Return']].head())
```

```python
stock['MA20'] = stock['Close'].rolling(20).mean()
stock['MA50'] = stock['Close'].rolling(50).mean()
```

```python
plt.figure(figsize=(12,6))
plt.plot(stock['Close'], label='Close', color='blue')
plt.plot(stock['MA20'], label='20-Day MA', color='red')
plt.plot(stock['MA50'], label='50-Day MA', color='green')
plt.title('Stock Price & Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()
plt.show()
```
